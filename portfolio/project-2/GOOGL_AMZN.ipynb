In this notebook I will be comparing the stock price history and standard deviation of Google, starting from Oct 1 2016 to Sep 30 2021. I will also compare Google against Amazon to find out patterns in data.

Data Background - The data is sourced from Yahoo finance and is in CSV format, with no null values or errors. This means the data can directly be used without any cleaning/transformation. 

I'll start by importing the `pandas` & `matplotlib` library and import the data for GOOGLE.

```python
import pandas as pd
import matplotlib.pyplot as plt

#brand color codes
g = '#DB4437'
a = '#FF9900'

googl = pd.read_csv("https://rahuldf.github.io/portfolio/project-2/GOOGL.csv")

googl.head()
```

With the `head()` function we can see the first five rows to the dataframe. For the first plot I will plot the Closing price of each day.

Below is the Stock Price history for Google from Oct 3, 2016 to Sep 30, 2021.



```python
plt.figure(num='GOOGLE Share History',figsize=(21,9))
plt.style.use('seaborn')
plt.plot(googl['Date'], googl['Close'], color=g)
plt.title("GOOGL Share Price | 2016-2021")
plt.xticks(googl['Date'][::140].tolist()+['2021-09-30'])
plt.show()
```

With the line graph above we can interpret that the stock was fairly stable at the start until Feb 2020 after which we see a dip due to the pendemic and a rise after that.

We can further find out the volatility by finding the difference between opening and closing price. I am adding a new column `Volatility` to the dataframe to hold this data.

Using predefined functions we can find out the standard deviation and mean of the daily Difference.

```python
googl['Volatility'] = googl['Open'] - googl['Close']

print("Standard Deviation = ",googl['Volatility'].std())
print("Mean = ",googl['Volatility'].mean())
print("Max Value = ",googl['Volatility'].max())
print("Min Value = ",googl['Volatility'].min())
```

We can see the Dtandard Deviation is 17.32 with a mean of -0.66. We can also plot this data on a histogram.

```python
plt.figure(num='GOOGLE Histogram', figsize=(21,9))
plt.hist(googl['Volatility'], bins=100, color=g)
plt.xlabel('Price Volatility in USD')
plt.ylabel('Count')
plt.show()
```

With the histogram plotted we can see the mean at around 0 and more data dispersed between -15 to +15. This also resembles a normal distribution curve.

I will now import the data for AMAZON and plot this against GOOGLE just to find patterns in the data.

```python
amzn = pd.read_csv("https://rahuldf.github.io/portfolio/project-2/AMZN.csv")

plt.figure(num='AMAZON vs GOOGLE Share History', figsize=(21,9))
plt.plot(amzn['Date'], amzn['Close'], color=a, label='AMAZON')
plt.plot(googl['Date'], googl['Close'], color=g, label='GOOGLE')
plt.title("AMAZON vs GOOGLE Share Price | 2016-2021")
plt.xticks(googl['Date'][::140].tolist()+['2021-09-30'])
plt.legend()
plt.show()
```

With both the stocks plotted together we can see that these started at around 800, with very similar patern thoroughout. However AMZN has been more aggressive after Oct 2017 and has managed to stay over GOOGL consistently.

This suggests the AMAZON has more deviation as compared to GOOGLE. Let us calculate the Standard Deviation for AMAZON to find out more.

```python
amzn['Volatility'] = amzn['Open'] - amzn['Close']

print("Standard Deviation = ",amzn['Volatility'].std())
print("Mean = ",amzn['Volatility'].mean())
print("Max Value = ",amzn['Volatility'].max())
print("Min Value = ",amzn['Volatility'].min())
```

With a Standard Deviation of 32.57 AMAZON is clearly more volatile then GOOGLE. 

We can also plot the data for both GOOGLE & AMAZON in a histogram to visualise the same.

```python
plt.figure(num='AMAZON vs GOOGLE Histogram', figsize=(21,9))
plt.hist(amzn['Volatility'], bins=100, color='#FF9900', label='AMAZON')
plt.hist(googl['Volatility'], bins=100, color='#DB4437', label='GOOGLE')
plt.xlabel('Price in USD')
plt.ylabel('Count')
plt.legend()
plt.show()
```

With both these stocks plotted together we can see AMZN being more volatile has major differences between -40 to +40 which is quite more then GOOGL's -15 to +15.

**Conclusion**

Stock prices as generally being volatile in nature do not carry same volatility. All stocks behave differently based on their respective industry and capacity, with some being more volatile then others. We can find out the volatility of these stocks with the help of statistics and make better investment decisions based on our risk capacity.
